```python
import matplotlib.pyplot as plt
years = [1950,
         1955,1960,1965,1970,1975,1980,1985,1990,1995,2000,
         2005,2010,2015]
pops = [2.5,2.7,3.0,3.3,3.6,
        4.0,4.4,4.8,5.3,5.7,6.1,6.5,6.9,7.3]
plt.plot(years,pops,color = (255/255,100/255,100/255))
plt.xlabel("Population in billions")
plt.ylabel("Population growth by year ")
plt.title("Population growth")
plt.show()


#it draws a     GRAPH  and add custom color and add label
```

```python
# now for multiple lines
import matplotlib.pyplot as plt
years = [1950,
         1955,1960,1965,1970,1975,1980,1985,1990,1995,2000,
         2005,2010,2015]
pops = [2.5,2.7,3.0,3.3,3.6,
        4.0,4.4,4.8,5.3,5.7,6.1,6.5,6.9,7.3]
deaths  = [1.2,1.7,1.8,2.2,2.5,
          2.7,2.9,3.1,3.3,3.5,3.8,4.0,4.3,4.4]
plt.plot(years,pops,color = (255/255,100/255,100/255))
plt.plot(years,deaths,"--",color = (.6,.6,1))
plt.xlabel("Population in billions")
plt.ylabel("Population growth by year ")
plt.title("Population growth")
plt.show()


#it draws a     GRAPH  and add custom color and add label
```

```python
# now for multiple lines
import matplotlib.pyplot as plt
years = [1950,
         1955,1960,1965,1970,1975,1980,1985,1990,1995,2000,
         2005,2010,2015]
pops = [2.5,2.7,3.0,3.3,3.6,
        4.0,4.4,4.8,5.3,5.7,6.1,6.5,6.9,7.3]
deaths  = [1.2,1.7,1.8,2.2,2.5,
          2.7,2.9,3.1,3.3,3.5,3.8,4.0,4.3,4.4]
lines = plt.plot(years,pops,years,deaths)
plt.grid(True)
plt.setp(lines,color=(1,.4,.4),marker ="+")
plt.show()


#it draws a     GRAPH  and add custom color and add label
```

```python
import matplotlib.pyplot as plt
labels = 'Python','C++','Ruby','Java','PHP','Perl'
separated=(1,0,0,0,0,)
sizes = [33,52,12,17,62,48]
plt.pie(sizes,labels =labels,autopct='%1.1f%%') #autopct to show percentages
plt.axis('equal') # in other place it  stretches so to keep it constant"
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = 'Python','C++','Ruby','Java','PHP','Perl'
separated=(.1,0,0,0,0,0) #HOW far you wqnt to separate it
sizes = [33,52,12,17,62,48]
plt.pie(sizes,labels =labels,autopct='%1.1f%%',explode =separated) #autopct to show percentages
plt.axis('equal') # in other place it  stretches so to keep it constant"
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
raw_data = {'names':['Nick','Panda','Azazel','lucifer','Belial'],
            'jan_ir':[143,122,101,106,360],
            'feb_ir':[122,122,135,651,64],
            'march_ir':[65,88,12,32,605]}
ddf =pd.DataFrame(raw_data)
ddf['total_ir'] = ddf['jan_ir'] + ddf['feb_ir'] + ddf['march_ir']
color =[(1,.4,.4),(1,.6,1),(.5,.3,1),(.3,1,.5),(.7,.7,.2)]

plt.pie(ddf['total_ir'],
        labels = ddf['names'],
        colors =color,
        autopct = '%1.1f%%')
plt.axis('equal')
plt.show()
```
