```python
import matplotlib.pyplot as plt
import numpy as py
col_count = 3
bar_width = .1
korea_scores =(554,536,538)
canada_scores = (518,523,525)
china_scores=(613,570,580)
france_scores=(495,505,489)
index = py.arange(col_count)
k1 = plt.bar(index,korea_scores,bar_width,alpha = .4,label="Korea")
c1 = plt.bar(index+0.2,canada_scores,bar_width,alpha = .4,label="Canada")
ch2 = plt.bar(index+0.4,china_scores,bar_width,alpha = .4,label="China")
f1 =plt.bar(index+.6,france_scores,bar_width,alpha = .4,label="France")
def createlabels(data):
    for item in data:
        height = item.get_height()
        plt.text(item.get_x() +item.get_width() /2,height*1.05,
               '%d' %int(height),
                ha='center',va = 'bottom')
plt.ylabel("Mean score in PISA 2021")
plt.xlabel("Subjects")
plt.title("Test score by countru")
createlabels(k1)
createlabels(c1)
createlabels(ch2)
createlabels(f1)

plt.xticks(index + .4/2,("Mathmatics","Reading","Science"))
plt.legend()
plt.grid(True)

plt.show()
```
